## **UKURAN 1024**

### RESIZE + konfigurasi anotasi

```python
import os
from PIL import Image
import pandas as pd

# Path folder gambar dan folder output
image_folders = [
    '/content/drive/MyDrive/210411100054-SitiNurAini/dataset/datasets_1024/Cardiomegaly/images',
    '/content/drive/MyDrive/210411100054-SitiNurAini/dataset/datasets_1024/Nodule_Mass/images',
    '/content/drive/MyDrive/210411100054-SitiNurAini/dataset/datasets_1024/Pneumothorax/images'
]

output_folders = [
    '/content/drive/MyDrive/210411100054-SitiNurAini/dataset/resize-1024-to-512/Cardiomegaly',
    '/content/drive/MyDrive/210411100054-SitiNurAini/dataset/resize-1024-to-512/Nodule_Mass',
    '/content/drive/MyDrive/210411100054-SitiNurAini/dataset/resize-1024-to-512/Pneumothorax'
]

# Pastikan folder output ada
for folder in output_folders:
    os.makedirs(folder, exist_ok=True)

# Load file CSV dengan bounding box
input_bbox_csv = '/content/drive/MyDrive/210411100054-SitiNurAini/dataset/datasets_1024/data.csv'
df = pd.read_csv(input_bbox_csv)

# Resize semua bounding box di CSV
resize_factor = 512 / 1024
df['x_min'] = (df['x_min'] * resize_factor).round().astype(int)
df['y_min'] = (df['y_min'] * resize_factor).round().astype(int)
df['x_max'] = (df['x_max'] * resize_factor).round().astype(int)
df['y_max'] = (df['y_max'] * resize_factor).round().astype(int)

# Membuat file CSV output baru dengan kolom `image_id` dan `class_name`
resized_bbox_csv = '/content/drive/MyDrive/210411100054-SitiNurAini/dataset/resize-1024-to-512/1024-to-512.csv'
df.to_csv(resized_bbox_csv, index=False)

# Resize semua gambar di setiap folder
for image_folder, output_folder in zip(image_folders, output_folders):
    for image_name in os.listdir(image_folder):
        input_path = os.path.join(image_folder, image_name)
        output_path = os.path.join(output_folder, image_name)

        # Resize gambar
        with Image.open(input_path) as img:
            img_resized = img.resize((512, 512))
            img_resized.save(output_path)

print("Proses resize gambar dan bounding box selesai!")
```

```python
import os
import pandas as pd
from xml.etree.ElementTree import Element, SubElement, tostring
import xml.dom.minidom

# Fungsi untuk membuat file XML dari data bounding box CSV
def create_xml_annotation(image_id, img_width, img_height, bounding_boxes, output_folder):
    annotation = Element('annotation')

    # Tambahkan elemen folder dan filename
    folder = SubElement(annotation, 'folder')
    folder.text = output_folder
    filename = SubElement(annotation, 'filename')
    filename.text = f"{image_id}.jpg"

    # Tambahkan elemen size
    size = SubElement(annotation, 'size')
    width = SubElement(size, 'width')
    width.text = str(img_width)
    height = SubElement(size, 'height')
    height.text = str(img_height)
    depth = SubElement(size, 'depth')
    depth.text = "1"  # Asumsi gambar berwarna (RGB)

    # Tambahkan setiap bounding box
    for bbox in bounding_boxes:
        class_id, x_min, y_min, x_max, y_max = bbox

        # Tambahkan elemen object
        obj = SubElement(annotation, 'object')
        name = SubElement(obj, 'name')
        name.text = class_id  # class_id harus berupa nama kelas
        pose = SubElement(obj, 'pose')
        pose.text = 'Unspecified'
        truncated = SubElement(obj, 'truncated')
        truncated.text = '0'
        difficult = SubElement(obj, 'difficult')
        difficult.text = '0'

        # Tambahkan elemen bndbox
        bndbox = SubElement(obj, 'bndbox')
        xmin = SubElement(bndbox, 'xmin')
        xmin.text = str(int(x_min))
        ymin = SubElement(bndbox, 'ymin')
        ymin.text = str(int(y_min))
        xmax = SubElement(bndbox, 'xmax')
        xmax.text = str(int(x_max))
        ymax = SubElement(bndbox, 'ymax')
        ymax.text = str(int(y_max))

    # Simpan file XML
    xml_str = xml.dom.minidom.parseString(tostring(annotation)).toprettyxml(indent="  ")
    xml_path = os.path.join(output_folder, f"{image_id}.xml")
    with open(xml_path, "w") as f:
        f.write(xml_str)
    print(f"File {xml_path} berhasil dibuat.")

# Contoh penggunaan
class_mapping = {
    'Cardiomegaly': 'Cardiomegaly',
    'Nodule/Mass': 'Nodule_Mass',
    'Pneumothorax': 'Pneumothorax'
}

# Baca file CSV
csv_file = '/content/drive/MyDrive/210411100054-SitiNurAini/dataset/resize-1024-to-512/1024-to-512.csv'
data = pd.read_csv(csv_file)
output_folder = '/content/drive/MyDrive/210411100054-SitiNurAini/DATASET-CXR-AL3/annotations/'

# Buat folder baru jika belum ada
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Ukuran gambar output yang sudah pasti 512x512
output_width = 512
output_height = 512

# Loop untuk memproses setiap baris
yolox_data = {}
for index, row in data.iterrows():
    image_id = row['image_id']
    class_name = row['class_name']
    x_min = row['x_min']
    y_min = row['y_min']
    x_max = row['x_max']
    y_max = row['y_max']

    # Mapping class name to ID
    class_id = class_mapping[class_name]

    # Simpan bounding box ke dictionary tanpa konversi
    if image_id not in yolox_data:
        yolox_data[image_id] = []
    yolox_data[image_id].append((class_id, x_min, y_min, x_max, y_max))

# Simpan file XML
for image_id, bounding_boxes in yolox_data.items():
    create_xml_annotation(image_id, output_width, output_height, bounding_boxes, output_folder)

print("Konversi ke XML selesai!")
```

```python
import os
import xml.etree.ElementTree as ET

# Path ke folder yang berisi file XML
xml_folder = '/content/drive/MyDrive/210411100054-SitiNurAini/DATASET-CXR-AL3/annotations/'

# Fungsi untuk memeriksa berapa banyak objek yang ada di dalam file XML
def cek_xml_untuk_objek_ganda(xml_path):
    tree = ET.parse(xml_path)
    root = tree.getroot()

    # Temukan semua elemen <object>
    objek = root.findall('object')

    # Jika ada lebih dari 1 objek, kembalikan True
    return len(objek) > 1

# Loop melalui semua file XML di folder yang ditentukan
xml_files_with_multiple_objects = []
for xml_file in os.listdir(xml_folder):
    if xml_file.endswith('.xml'):
        xml_path = os.path.join(xml_folder, xml_file)

        # Periksa apakah file XML mengandung lebih dari 1 objek
        if cek_xml_untuk_objek_ganda(xml_path):
            xml_files_with_multiple_objects.append(xml_file)

# Tampilkan jumlah dan file XML yang mengandung lebih dari 1 objek
if xml_files_with_multiple_objects:
    print(f"Jumlah file XML yang mengandung lebih dari 1 objek: {len(xml_files_with_multiple_objects)}")
    print("File XML yang mengandung lebih dari 1 objek:")
    for xml_file in xml_files_with_multiple_objects:
        print(xml_file)
else:
    print("Tidak ada file XML yang mengandung lebih dari 1 objek ditemukan.")
```

### **RENAME 512 (supaya bisa digabung dengan data folder 1024) KARENA PENAMAAN FILE NYA SAMA**

```python
import os
import shutil

# Folder input
image_folder = '/content/drive/MyDrive/210411100054-SitiNurAini/dataset/DATA-512/image'
label_folder = '/content/drive/MyDrive/210411100054-SitiNurAini/dataset/DATA-512/anotasi'

# Folder output
output_image_folder = '/content/drive/MyDrive/210411100054-SitiNurAini/dataset/RENAME-DATA-512/image'
output_label_folder = '/content/drive/MyDrive/210411100054-SitiNurAini/dataset/RENAME-DATA-512/anotasi'

# Membuat folder output jika belum ada
os.makedirs(output_image_folder, exist_ok=True)
os.makedirs(output_label_folder, exist_ok=True)

# Fungsi untuk mengganti nama file gambar dan label
def rename_and_copy_files(image_folder, label_folder, output_image_folder, output_label_folder):
    image_files = sorted(os.listdir(image_folder))  # Mengurutkan file gambar

    for filename in image_files:
        # Mendapatkan nama dan ekstensi file
        name, ext = os.path.splitext(filename)
        new_name = f"{name}_512{ext}"  # Tambahkan _512 sebelum ekstensi

        old_image_path = os.path.join(image_folder, filename)
        new_image_path = os.path.join(output_image_folder, new_name)

        old_label_name = f"{name}.xml"  # Nama file label terkait
        old_label_path = os.path.join(label_folder, old_label_name)
        new_label_name = f"{name}_512.xml"  # Nama label baru
        new_label_path = os.path.join(output_label_folder, new_label_name)

        # Menyalin gambar
        shutil.copy(old_image_path, new_image_path)

        # Menyalin label jika ada
        if os.path.exists(old_label_path):
            shutil.copy(old_label_path, new_label_path)

# Memanggil fungsi
rename_and_copy_files(image_folder, label_folder, output_image_folder, output_label_folder)
print("Proses rename dan copy selesai.")
```
